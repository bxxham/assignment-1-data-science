```python
# BSE-05-0132/2024.
# BECKHAM MWIRIGI.
# GROUP 2.

import matplotlib.pyplot as plt
import numpy as np

values1 = [50, 53, 59, 56, 62]
values2 = [39, 47, 42, 51, 51]
values3 = [70, 80, 90, 87, 92]
y = ["07/2019", "08/2019", "09/2019", "10/2019", "11/2019"]

x = np.arange(len(y))
width = 0.25

plt.bar(x - width,  values1, width, color = 'blue', label = "Searches")
plt.bar(x, values2, width, color = 'hotpink', label = "Direct")
plt.bar(x + width,  values3, width, color = 'orange', label = "Social Media")

plt.xlabel("Months")
plt.ylabel("Visitors in Thousands")
plt.title("Visitors by Web Traffic Sources")
plt.xticks(x, y)
plt.legend()

plt.show()
```
